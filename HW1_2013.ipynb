```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
from pattern import web
from fnmatch import fnmatch


# set some nicer defaults for matplotlib
from matplotlib import rcParams
```

```python
from urllib.request import urlopen
#your code here    
def get_poll_xml(poll_id):
    link = 'http://charts.realclearpolitics.com/charts/'+ str(poll_id) +'.xml'
    xml = requests.get(link).text
    return xml
    
xml = get_poll_xml(1171)
```

```python
def rcp_poll_data1(xml):
    dom = web.Element(xml)
    dates = dom.by_tag('series')[0]
    dates = {n.attributes['xid']:str(n.content) for n in dates.by_tag('value')}
    keys = dates.keys()
    result= {}
    result['Dates'] = pd.to_datetime([dates[k] for k in keys])
    for graph in dom.by_tag('graph'):
        title = graph.attributes['title']
        data = {v.attributes['xid']:float(v.content) if v.content else np.nan for v in graph.by_tag('value')}
        result[title] = [data[k] for k in keys]
    df = pd.DataFrame(result)
    return df
rcp_poll_data1(xml)
```

```python
html = requests.get('https://www.realclearpolitics.com/epolls/2010/governor/2010_elections_governor_map.html').text
def find_governor_races(html):
    dom = web.Element(html)
    pattern = 'http://www.realclearpolitics.com/epolls/????/governor/??/*-*.html'
    links = [a.attributes.get('href',"") for a in dom.by_tag('a')]
    links = [l for l in links if fnmatch(l,pattern)]
    links = list(set(links))
    return links
find_governor_races(html)
```
