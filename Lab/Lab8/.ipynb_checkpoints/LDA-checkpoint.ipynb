```python
import numpy as np
import pandas as pd
import scipy as sp
from sklearn import linear_model
import matplotlib
import matplotlib.pyplot as plt
from sklearn import discriminant_analysis as da
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
fig,ax = plt.subplots(2,3,figsize = (12,5))
ax = ax.ravel()

for axs,i in enumerate(range(0,11,2)):
    data = np.loadtxt('dataset_1_year_'+str(2000+i)+'.txt')
    
    x = data[:,:-1]
    y = data[:,-1]
    
    ax[axs].scatter(x[y==2,0],x[y==2,1],color = 'green',label = '2-High income', alpha=0.4)
    ax[axs].scatter(x[y==1,0],x[y==1,1],color = 'blue',label = '1-Middle income', alpha=0.4)
    ax[axs].scatter(x[y==0,0],x[y==0,1],color = 'red',label = '0-Low income', alpha=0.4)
    
    ax[axs].set_xlabel('Latitude')
    ax[axs].set_ylabel('Longitude')
    ax[axs].set_title(str(2000+i))
plt.legend(bbox_to_anchor=(1,1),loc='upper left', ncol=1)
plt.tight_layout()
```

```python
data = np.loadtxt('dataset_1_year_2000.txt')

x = data[:,:-1]
y = data[:,-1]

logreg = linear_model.LogisticRegression(C=1000,fit_intercept=True)
logreg.fit(x,y)

lda = da.LinearDiscriminantAnalysis()
lda.fit(x,y)

knn = KNN(n_neighbors=5)
knn.fit(x,y)

logregtrain = logreg.score(x,y)
ldrtrain = lda.score(x,y)
knntrain = knn.score(x,y)

acc_log = []
acc_lda = []
acc_knn = []

for i in range(1, 11):
    #Load data for year 2000 + i 
    data_i = np.loadtxt('dataset_1_year_' + str(2000 + i) + '.txt')
    
    #Split predictors, response
    x_i = data_i[:, :-1]
    y_i = data_i[:, -1]

    #Compute predictive accuracies of various models trained on 2000 data
    acc_log.append(logreg.score(x_i, y_i))
    acc_lda.append(lda.score(x_i, y_i))
    acc_knn.append(knn.score(x_i, y_i))

plt.figure(figsize=(10,7))
plt.bar(np.arange(1,11)+2000,acc_log,width = 0.25,color = 'blue',alpha = 0.5,label = 'LogReg')
plt.bar(np.arange(1,11)+2000+0.25,acc_lda,width = 0.25,color = 'red',alpha = 0.5,label = 'LDA')
plt.bar(np.arange(1,11)+2000+0.25+0.25,acc_knn,width = 0.25,color = 'green',alpha = 0.5,label = 'KNN')

plt.axhline(logregtrain,color='blue',label = 'LogReg - 2000')
plt.axhline(knntrain,color = 'green',label = 'KNN - 2000')
plt.axhline(ldrtrain,color = 'red',label = 'LDA - 2000')
plt.legend(loc = 'best',bbox_to_anchor=(1,1))
```

```python
#Visualize data for every other year
fig, ax = plt.subplots(3, 6, figsize = (23, 10))

#Iterate from year 0 to 10, with steps of 2
for i in range(0, 11, 2):
    #Load data
    data = np.loadtxt('dataset_1_year_' + str(2000 + i) + '.txt')
    
    #Split into predictor/response
    x = data[:, :-1]
    y = data[:, -1]
    
    
    #Plot each class for the current year in different colors
    ax[0, i // 2].scatter(x[y == 2, 0], x[y == 2, 1],
                         color='g', alpha=0.6,
                         label = '2-High income')
    ax[1, i // 2].scatter(x[y == 1, 0], x[y == 1 ,1], 
                         color='b', alpha=0.6,
                         label = '1-Middle class')
    ax[2, i // 2].scatter(x[y == 0, 0], x[y == 0, 1], 
                         color='r', alpha=0.6,
                         label = '0-Low income')
    
    #Labels
    ax[0, i // 2].set_xlabel('Latitude')
    ax[0, i // 2].set_ylabel('Longitude')
    ax[0, i // 2].set_title(str(2000 + i))
    
    #Labels
    ax[1, i // 2].set_xlabel('Latitude')
    ax[1, i // 2].set_ylabel('Longitude')
    ax[1, i // 2].set_title(str(2000 + i))
    
    # LABEL AXIS, TITLE
    ax[2, i // 2].set_xlabel('Latitude')
    ax[2, i // 2].set_ylabel('Longitude')
    ax[2, i // 2].set_title(str(2000 + i))


plt.legend(bbox_to_anchor=(1, 1), loc='upper left', ncol=1)
plt.tight_layout()
```

```python
data = np.loadtxt('dataset_1_year_2000.txt')

x = data[:,:-1]
y = data[:,-1]

lda = da.LinearDiscriminantAnalysis()
lda.fit(x,y)

ldatrain = lda.score(x,y)

acc_lda_corrected = []

#Store class proportions in 2000
p0_2000 = (y==0).mean()
p1_2000 = (y==1).mean()
p2_2000 = (y==2).mean()

#Store class proportions in current year
p0_current = (y==0).mean()
p1_current = (y==1).mean()
p2_current = (y==2).mean()

for k in range(1,11):
    data = np.loadtxt('dataset_1_year_'+str(2000+k)+'.txt')
    
    x_i = data[:,:-1]
    y_i = data[:,-1]
    
    p2_current = p2_current*1.25
    p0_current = (1-p2_current)/2
    p1_current = (1-p1_current)/2
    
    pred_logprob = lda.predict_log_proba(x_i) # compute log-probabilities
    pred_logprob[:, 0] = pred_logprob[:, 0] + np.log(p0_current / p0_2000) # correction for class 0
    pred_logprob[:, 1] = pred_logprob[:, 1] + np.log(p1_current / p1_2000) # correction for class 1
    pred_logprob[:, 2] = pred_logprob[:, 2] + np.log(p2_current / p2_2000) # correction for class 2

    
    #Predict class label using re-estimated probabilities
    y_pred = pred_logprob.argmax(axis = 1)
    
    #Compute accuracy 
    acc_lda_corrected.append(np.mean(y_i == y_pred))
    
#Plot accuracy over years
years = np.arange(1, 11) + 2000 # x-axis is years 2001-2010
width = 0.25 # width of bar

plt.figure(figsize=(10,3))
plt.bar(years + width, acc_lda, width, color='red', alpha=0.5, label='LDA')
plt.bar(years + 2 * width, acc_lda_corrected, width, color='green', alpha=0.5, label='LDA yearly')

plt.xlabel('Year')
plt.ylabel('Prediction Accuracy')
plt.legend(loc = 'upper left', bbox_to_anchor=(1, 1)) # legend upper left outside
plt.title('Overall predictive accuracy of LDA models')
plt.xticks(years + width + 0.125, years); # set x-ticks spacing
plt.axhline(ldatrain, 0, 1, color="red", label="LDA 2000")
```

```python
data = np.loadtxt('dataset_2.txt')

x = data[:,:-1]
y = data[:,-1]

fig,ax = plt.subplots(1,1,figsize = (5,5))
model = linear_model.LogisticRegression()
model.fit(x,y)
acc_logreg = model.score(x,y)

ax.scatter(x[y==1,0],x[y==1,1],c='red',cmap=plt.cm.coolwarm, alpha=0.2, s=20)
ax.scatter(x[y==0,0],x[y==0,1],c='blue',cmap=plt.cm.coolwarm, alpha=0.2, s=20)

interval = np.arange(0,1,0.01)
n = np.size(interval)

x1,x2 = np.meshgrid(interval,interval)
x1 = x1.reshape(-1,1)
x2 = x2.reshape(-1,1)
xx =np.concatenate((x1,x2),axis = 1)
yy = model.predict(xx)
yy = yy.reshape((n,n))
x1 = x1.reshape(n,n)
x2 = x2.reshape(n,n)
ax.contourf(x1,x2,yy,alpha=0.2,cmap=plt.cm.coolwarm)
```
